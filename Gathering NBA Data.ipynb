```python
import pandas as pd
from nba_api.stats.endpoints import leaguegamelog
from nba_api.stats.endpoints import boxscoretraditionalv2
```

```python
seasons = list(range(2011,2021))
```

```python
regular_games = leaguegamelog.LeagueGameLog(season = 2010, season_type_all_star = "Regular Season")
regular_games = regular_games.get_data_frames()[0]
for s in seasons:
    dat = leaguegamelog.LeagueGameLog(season = s, season_type_all_star = "Regular Season")
    dat = dat.get_data_frames()[0]
    regular_games = pd.concat([regular_games, dat])
```

```python
regular_games
```

```python
playoff_games = leaguegamelog.LeagueGameLog(season = 2010, season_type_all_star = "Playoffs")
playoff_games = playoff_games.get_data_frames()[0]
for s in seasons:
    dat = leaguegamelog.LeagueGameLog(season = s, season_type_all_star = "Playoffs")
    dat = dat.get_data_frames()[0]
    playoff_games = pd.concat([playoff_games, dat])
```

```python
playoff_games
```

```python
teams = dat.drop_duplicates(subset='TEAM_ID', keep="last")
teams = teams[["TEAM_ID", "TEAM_NAME"]]
teams = teams.reset_index(drop = True)
teams
```

```python
reg_ids = regular_games["GAME_ID"].unique()
```

```python
bs_trad_regular = boxscoretraditionalv2.BoxScoreTraditionalV2(game_id = reg_ids[0])
bs_trad_regular = bs_trad_regular.get_data_frames()[0]
for i in range(1, len(reg_ids)):
    dat = boxscoretraditionalv2.BoxScoreTraditionalV2(game_id = reg_ids[i])
    dat = dat.get_data_frames()[0]
    bs_trad_regular = pd.concat([bs_trad_regular, dat])
```

```python
bs_trad_regular
```

```python
playoff_ids = playoff_games["GAME_ID"].unique()
```

```python
bs_trad_playoff = boxscoretraditionalv2.BoxScoreTraditionalV2(game_id = playoff_ids[0])
bs_trad_playoff = bs_trad_playoff.get_data_frames()[0]
for i in range(1, len(playoff_ids)):
    dat = boxscoretraditionalv2.BoxScoreTraditionalV2(game_id = playoff_ids[i])
    dat = dat.get_data_frames()[0]
    bs_trad_playoff = pd.concat([bs_trad_playoff, dat])
```
